# **⭐ Classification Model to Identify Multiple Disease**

```python
# import library
import pandas as  pd
```

```python
# import data
disease = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/MultipleDiseasePrediction.csv')
```

```python
# view data
disease.head()
```

```python
# info of data
disease.info(verbose = True , null_counts = True)
```

```python
# summary statistics
disease.describe()
```

```python
# check for missing value
disease.isnull().sum()
```

```python
# check for categories
disease.nunique()
```

```python
# correlation
disease.corr()
```

```python
# column names
disease.columns
```

```python
# define y
y = disease['prognosis']
```

```python
# define X
x = disease[['itching', 'skin_rash', 'nodal_skin_eruptions', 'continuous_sneezing',
       'shivering', 'chills', 'joint_pain', 'stomach_pain', 'acidity',
       'ulcers_on_tongue',
       'blackheads', 'scurring', 'skin_peeling', 'silver_like_dusting',
       'small_dents_in_nails', 'inflammatory_nails', 'blister',
       'red_sore_around_nose', 'yellow_crust_ooze']]
```

```python
# split data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, random_state = 2529)
```

```python
# verify shape
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# select model
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
# train model
model.fit(x_train, y_train)
```

```python
# predict with model
y_pred = model.predict(x_test)
```

```python
# model evaluation
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# model accuracy
accuracy_score(y_test, y_pred)
```

```python
# model confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
# model classification report
print(classification_report(y_test, y_pred))
```
